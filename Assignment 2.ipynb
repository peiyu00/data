### NumPy

```python
import numpy as np
```

```python
# Task 1
path = "Attendance_RAW.txt"
lines = [x.rstrip() for x in open(path)]

attend = np.array(lines, dtype=int)
attend
```

```python
# Task 2
attend = attend.reshape(14,15).T
attend
```

```python
# Task 3
def average(record):
    avg = record.mean(axis=0).round(2)
    return avg
    
attend_per_day = average(attend)
attend_per_day
```

### Pandas

```python
import pandas as pd
```

```python
# Task 1
emp_detail = pd.read_csv("emp_detail.csv")
```

```python
# Task 2
emp_detail.head()
```

```python
# Task 3a
attendance = pd.DataFrame(attend)
attendance
```

```python
# Task 3b
emp_attendance = pd.concat([emp_detail, attendance], axis = 1)
emp_attendance
```

```python
# Task 4
emp_attendance['total'] = attendance.sum(axis = 1)
emp_attendance
```

```python
# Task 5
emp_attendance['rate'] = round((emp_attendance['total']/14)*100, 2)
emp_attendance
```

```python
# Task 6
emp_attendance['status'] = np.where(emp_attendance['rate']>90, "Good", np.where(emp_attendance['rate']>60, "Moderate", "Bad"))
emp_attendance
```

```python
# Task 7
def department_avg(record):
    avg = record.groupby('Department').mean()
    return avg['total']
    
attend_by_depart = department_avg(emp_attendance)
attend_by_depart
```
